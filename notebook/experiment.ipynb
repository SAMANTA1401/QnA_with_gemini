```python
documents=SimpleDirectoryReader("../Data")
```

```python

doc=documents.load_data()
```

```python

print(doc[0].text)
```

```python
model=Gemini(models='gemini-pro',api_key=google_api_key)
```

```python
gemini_embed_model=GeminiEmbedding(model_name="models/embedding-001")
```

```python
import llama_index
from llama_index.core import Settings
from llama_index.llms.gemini import Gemini
from llama_index.core.node_parser import SentenceSplitter
```

```python
Settings.llm = Gemini(model='models/gemini-pro')
Settings.embed_model = GeminiEmbedding(model="models/embedding-001")
Settings.node_parser = SentenceSplitter(chunk_size=800, chunk_overlap=20)
# Settings.num_output = 512
# Settings.context_window = 3900
```

```python
# a vector store index only needs an embed model
index = VectorStoreIndex.from_documents(doc, embed_model=gemini_embed_model)
```

```python
index
```

```python

index.storage_context.persist()
```

```python
query_engine=index.as_query_engine()
```
